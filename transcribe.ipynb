```python
import os
import whisper
from pathlib import Path
from utils import save_txt

testset_dir = Path("./testset")
if not testset_dir.exists():
    print("'testset' folder not found. Create it and add subfolders with MP3 files.")
    exit(1)

model = whisper.load_model("base")  

for subdir in testset_dir.iterdir():
    if not subdir.is_dir():
        continue  
    
    print(f"\nProcessing folder: {subdir.name}")
    
    feedback_files = list(subdir.glob("*_feedback.mp3")) 
    
    if not feedback_files:
        print(f"No *_feedback.mp3 files found in {subdir.name}")
        continue
    
    for feedback_file in feedback_files:
        print(f"   Transcribing: {feedback_file.name}")
        result = model.transcribe(str(feedback_file))
        transcription = result["text"]  
        
        output_file1 = subdir / "human_feedback.txt"
        human_feedback_dir = testset_dir / "human_feedback"
        output_file2 = human_feedback_dir / f"{subdir.name}.txt"  
        
        # Save to both
        save_txt(transcription, str(output_file1))
        save_txt(transcription, str(output_file2))
    
```

```python
import os

BASE_DIR = "./testset"

PREFIXED_SUFFIXES = [
    "part_1.mp3",
    "part_2.mp3",
    "part_3.mp3",
    "feedback.mp3",
]

NON_PREFIXED_FILES = [
    "human_feedback.txt",
]

for subdir in os.listdir(BASE_DIR):
    subdir_path = os.path.join(BASE_DIR, subdir)
    if not os.path.isdir(subdir_path):
        continue

    missing = []

    # Check prefixed files
    for suffix in PREFIXED_SUFFIXES:
        filename = f"{subdir}_{suffix}"
        if not os.path.exists(os.path.join(subdir_path, filename)):
            missing.append(filename)

    # Check non-prefixed files
    for filename in NON_PREFIXED_FILES:
        if not os.path.exists(os.path.join(subdir_path, filename)):
            missing.append(filename)

    if missing:
        print(f"\nMissing in folder: {subdir}")
        for m in missing:
            print(f"   - {m}")
    else:
        print(f"All files present in: {subdir}")
```
